# 연습_filter , mask

filter :특정 컬럼 선택 및 데이터 추출  
mask : 조건에 해당하는 데이터 행 지우기 

```python
#연습데이터 생성
df= pd.DataFrame({'ABC':['A','B','C','A'],
                 'abc':['a','b','c','c'],
                 'group1':['G1','G2','G3','G4'],
                 'value1':[11,13,12,14],
                 'val1_mod':[0.231,0.197,0.213,0.199]})
df
```

#### filter

```python
# 컬럼이름 직접입력 
df.filter(items=['group1','value1'])
```

```python
df
```

```python
#regex(regular express)이용, '1'포함
df.filter(regex='1',axis=1)
```

```python
#regex이용, '1'로 끝나는 컬럼
df.filter(regex='1$',axis=1)
```

```python
#'a'포함된 컬럼.regex = 'a' 와 동일
df.filter(like='a',axis =1)
```

```python
# 'A' 또는 'a' 포함된 컬럼([]이용)
df.filter(regex='[Aa]', axis=1)
```

#### mask

```python
# value1이 12 초과인 행 mask처리  
df.mask(df.value1 > 12)
```

```python
# abc 컬럼이 'a' 인 행 mask처리 
df.mask(df.abc == 'a')
```

```python
# value1이 12이상인 행을 999로 
df.mask(cond = df.value1>12,other=999)
```

# 데이터결합

concat : concatenate 연결   
append 추가 

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']},
                  index = [0,1,2,3])

df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                   'B':['B4','B5','B6','B7'],
                   'C':['C4','C5','C6','C7'],
                   'D':['D4','D5','D6','D7']},
                  index = [4,5,6,7])

df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                   'B':['B8','B9','B10','B11'],
                   'C':['C8','C9','C10','C11'],
                   'D':['D8','D9','D10','D11']},
                  index = [8,9,10,11])
```

```python
# concat => df1,df2,df3 결합 
result = pd.concat([df1,df2,df3])
result
```

```python
# 가로결합 : axis = 1 
result = pd.concat([df1,df2,df3],axis = 1)
result
```

```python
# ignore_index => 기존의 index가 아닌 새로운 index로 생성
result = pd.concat([df1,df2,df3],axis =1, ignore_index = True)
result
```

```python
# col_변수명 으로 변경
result.columns = ["col_" + str(i) for i in range(12)]
result 
```

# 데이터 결합

```python
left = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                   'key2':['K0','K1','K0','K1'],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})

right = pd.DataFrame({'key1':['K0','K1','K2','K3'],
                   'key2':['K0','K0','K0','K0'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
#left 와 right의 key1,key2컬럼의 값이 각각 같은 행만 출력 
result = pd.merge(left,right,on=['key1','key2'])
result
```

```python
#left 와 right의 key1,key2컬럼의 left 기준으로 출력, right값의 빈부분은 NaN처리  
result = pd.merge(left,right,on=['key1','key2'],how='left')
result
```

```python
#left 와 right의 key1,key2컬럼의 right 기준으로 출력, left값의 빈부분은 NaN처리  
result = pd.merge(left,right,on=['key1','key2'],how = 'right')
result
```

```python
# outer :합집합을 의미 , indicator=True 값을 주어 _merge 생성
result = pd.merge(left,right,on=['key1','key2'],how='outer',indicator=True)
result
```

# 연습

```python
left = pd.DataFrame({'key1':[1,2,3,4],
                   'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']})

right = pd.DataFrame({'C':['C1','C2','C3'],
                   'D':['D1','D2','D5']},
                    index=[1,2,5])
```

```python
#left에 right를 join => left기준으로 right의 열 추가 
result = left.join(right)
result
```

```python
#how = 'right'=> right기준으로 left의 열 추가 
result = left.join(right,how='right')
result
```

```python
left
```

```python
right
```

```python
result = left.join(right,how='inner')
result
```

```python
result = left.join(right,how='outer')
result
```

```python
result = left.join(right,on='key1')   # 특정 열 key1 기준으로 , on ='key1' 없을경우 index기준
result
```

# crosstab 

```python
df = pd.DataFrame({'factor1':['A','A','A','A','A','B','B','B','B'],
                  'factor2': ['a','b','c','a','b','a','b','c','c'],
                  'value':[11,13,12,12,14,12,13,11,12]})

df
```

```python
pd.crosstab(df.factor1,df.factor2)
```

```python
pd.crosstab(df.factor1,df.factor2,
           rownames = ['Fact1'],colnames=['Fact2'],
           margins=True)
```

```python
pd.crosstab(df.factor1,[df.factor2,df.value])
```

```python
#각 셀의 빈도 기준 정규화 확률 산출
pd.crosstab(df.factor1,df.factor2,
           margins=True, normalize =True)
```
